## Data Visualization

#### Population Growth:

How has the average population in South Africa changed over the years?

```python
import matplotlib.pyplot as plt
# Plotting population growth over the years
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['Average_Population'], marker='o', linestyle='-')
plt.title('Population Growth in South Africa (2002-2022)')
plt.xlabel('Year')
plt.ylabel('Average Population')
plt.grid(True)
plt.show()
```

**Observation:**
- South Africa's population has been steadily increasing from 2002 to 2022.
- There appears to be consistent growth with some fluctuations over the years.
- The growth seems to be linear, suggesting a gradual increase in population.

 #### Birth and Death Rates:

How have crude birth rates and crude death rates changed over time?

```python
# Plotting birth and death rates over the years
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['Crude_Birth_Rate'], marker='o', linestyle='-', label='Birth Rate')
plt.plot(df['Year'], df['Crude_Death_Rate'], marker='o', linestyle='-', label='Death Rate')
plt.title('Crude Birth and Death Rates in South Africa (2002-2022)')
plt.xlabel('Year')
plt.ylabel('Rate (per 1,000)')
plt.legend()
plt.grid(True)
plt.show()
```

**Observation:**
- The crude birth rate (births per 1,000 people) has been on a declining trend over the years.
- Conversely, the crude death rate (deaths per 1,000 people) has also been decreasing but at a slower rate.
- There has been a gradual decrease in the gap between birth and death rates, indicating a narrowing natural population increase.

#### Fertility Rates:

How have total fertility rates (TFR) evolved over the years?

```python
# Plotting total fertility rates (TFR) over the years
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['Total_Fertility_Rates'], marker='o', linestyle='-')
plt.title('Total Fertility Rates in South Africa (2002-2022)')
plt.xlabel('Year')
plt.ylabel('TFR')
plt.grid(True)
plt.show()
```

- Total fertility rates (TFR) have experienced a decline from around 2.45 in 2002 to approximately 2.34 in 2022.
- TFR has remained relatively stable with slight variations over the years.
- The TFR indicates the average number of children a woman is expected to have during her lifetime, and the data suggests a moderate decrease in fertility in South Africa.
